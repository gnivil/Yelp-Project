```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading in File from Census.gov CSV keeping only relevant rows out of the 60 available
pred = lambda x: x not in [0, 1, 27, 32, 45]
census_data_df = pd.read_csv("QuickFacts_Aug-06-2021.csv", skiprows=pred, index_col=0)
census_data_df
```

```python
# dropping unnecessary columns
census_data_df = census_data_df.drop(["Fact Note", "Value Note for Lake Forest city, Illinois",
                                     "Value Note for Mundelein village, Illinois", "Value Note for Lincolnshire village, Illinois",
                                     "Value Note for Libertyville village, Illinois", "Value Note for Vernon Hills village, Illinois",
                                     "Value Note for Lake County, Illinois","Lake County, Illinois"], axis=1)
```

```python
census_data_df
```

```python
census_data_df
```

```python
#transcribing data to allow for cleaner look and ploting
transcribe_census_df = census_data_df.T
transcribe_census_df  = transcribe_census_df .rename(columns={"Households with a broadband Internet subscription, percent, 2015-2019":"Households with a broadband (2019)"})
transcribe_census_df  = transcribe_census_df .rename(columns={"Median household income (in 2019 dollars), 2015-2019":"Median household income (2019)"})
```

```python
transcribe_census_df 
```

```python
#adjusting formating to change to float for plotting
transcribe_census_df['Households, 2015-2019'] = transcribe_census_df ['Households, 2015-2019'].str.replace(',','').astype(float)
transcribe_census_df['Households with a broadband (2019)'] = transcribe_census_df ['Households with a broadband (2019)'].str.replace('%','').astype(float)
```

```python
transcribe_census_df.dtypes
```

```python
# Get broadband info
broadband = transcribe_census_df['Households with a broadband (2019)']

# Configure plot and ticks
broadband.plot(kind="bar", facecolor="red")

# Set a Title and labels
plt.title("% of Households with Broadband")
plt.xlabel("City")
plt.ylabel("Household Percentage with Broadband ")
plt.tight_layout()
plt.show()
```

```python
# Get household info
household = transcribe_census_df['Households, 2015-2019']

# Configure plot and ticks
household.plot(kind="bar", facecolor="blue")

# Set a Title and labels
plt.title("Households by City")
plt.xlabel("City")
plt.ylabel("Households")
plt.tight_layout()
plt.show()
```

```python
#fixing formatting
transcribe_census_df['Median household income (2019)'] = transcribe_census_df ['Median household income (2019)'].str.replace(',','')
```

```python
#fixing formatting
transcribe_census_df['Median household income (2019)'] = transcribe_census_df['Median household income (2019)'].str.replace('$','').astype(float)
```

```python
# Get mean income info
med_income = transcribe_census_df['Median household income (2019)']

# Configure plot and ticks
med_income.plot(kind="bar", facecolor="green")

# Set a Title and labels
plt.title("Median Household income by City")
plt.xlabel("City")
plt.ylabel("Households")
plt.tight_layout()
plt.show()
```
